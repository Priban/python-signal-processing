```python
from wfdb import rdsamp, rdrecord

record = rdrecord('drive01', pn_dir="drivedb")
num_samples = record.sig_len

# Calculate the sampto value as the number of samples - 1
sampto = num_samples - 1

# Download the data using the specified sampling rate
data, fields = rdsamp('drive01', pn_dir="drivedb", sampfrom=0, sampto=sampto)
```

```python
wfdb.__version__
```

```python
import matplotlib.pyplot as plt

first_channel = data[0:60000, 0]

# Get the sampling frequency
sampling_frequency = fields['fs']

# Generate the time values for the x-axis
time = [i / sampling_frequency for i in range(len(first_channel))]

# Create a line plot of the data
plt.plot(time, first_channel)

# Limit the x-axis range to the first 5 seconds
plt.xlim(0, 5)

# Add labels and title
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('First channel of drive01 record')

# Show the plot
plt.show()
```
