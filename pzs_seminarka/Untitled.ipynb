```python
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
import random
from matplotlib.pyplot import figure
```

```python
figure(figsize=(20, 6), dpi=400)

a_nosny = random.random() * 5
f_nosny = random.random() * 50
delka = np.pi

a_mod = random.random() * 5
f_mod = random.random() * 300 + 100

x = np.linspace(0, delka, 1000)
sig = a_nosny * np.sin(x * f_nosny)
mod = a_mod * np.cos(x * f_mod)

plt.plot(x, sig, alpha=1)
plt.plot(x, mod, alpha=0.3)
plt.plot(x, sig + mod)
plt.show()
```
